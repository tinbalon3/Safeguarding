### LLM Guard là gì?
LLM Guard là khái niệm chỉ việc thêm các "biện pháp bảo vệ" (guardrails) vào các mô hình ngôn ngữ lớn (Large Language Models - LLMs) để đảm bảo an toàn, chính xác và phù hợp trong ứng dụng thực tế. Các guardrails này giúp:
- **Kiểm soát đầu vào**: Ngăn chặn các truy vấn độc hại hoặc không phù hợp.
- **Kiểm soát đầu ra**: Đảm bảo câu trả lời không chứa nội dung độc hại, hoặc không liên quan.

### Create a Guard from the installed guardrail.

```python
from guardrails import Guard, OnFailAction
from guardrails.hub import RegexMatch

guard = Guard().use(
    RegexMatch, regex="\(?\d{3}\)?-? *\d{3}-? *-?\d{4}", on_fail=OnFailAction.EXCEPTION
)

guard.validate("123-456-7890")  # Guardrail passes

try:
    guard.validate("1234-789-0000")  # Guardrail fails
except Exception as e:
    print(e)
```

#### Output

Validation failed for field with errors: Result must match \(?\d{3}\)?-? *\d{3}-? *-?\d{4}

### Run multiple guardrails within a Guard. First, install the necessary guardrails from Guardrails Hub.

```python
!guardrails hub install hub://guardrails/competitor_check
!guardrails hub install hub://guardrails/toxic_language
```

```python
from guardrails import Guard, OnFailAction
from guardrails.hub import CompetitorCheck, ToxicLanguage

guard = Guard().use_many(
    CompetitorCheck(["Apple", "Microsoft", "Google"], on_fail=OnFailAction.EXCEPTION),
    ToxicLanguage(threshold=0.5, validation_method="sentence", on_fail=OnFailAction.EXCEPTION)
)

guard.validate(
    """An apple a day keeps a doctor away.
    This is good advice for keeping your health."""
)  # Both the guardrails pass

try:
    guard.validate(
        """Shut the hell up! Apple just released a new iPhone."""
    )  # Both the guardrails fail
except Exception as e:
    print(e)
```

#### Output

Validation failed for field with errors: Found the following competitors: [['Apple']]. Please avoid naming those competitors next time, The following sentences in your response were found to be toxic:

- Shut the hell up!

### Kết nối Guardrails với LLM
Chúng ta sẽ dùng OpenAI như một ví dụ để tích hợp Guardrails, kiểm soát cả đầu vào và đầu ra.

#### Thiết lập OpenAI
Thêm API key vào biến môi trường:

```python
# Cài OpenAI
!pip install openai
```

```python
import os
from openai import OpenAI

# Thiết lập API key
os.environ["OPENAI_API_KEY"] = "your-openai-api-key"
client = OpenAI()

# Tạo hàm gọi OpenAI với Guardrails
def guarded_openai_call(prompt):
    # Kiểm tra đầu vào
    try:
        guard.validate(prompt)
        print("Đầu vào hợp lệ!")
    except Exception as e:
        return f"Đầu vào bị chặn: {e}"

    # Gọi OpenAI
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=100
    )
    output = response.choices[0].message.content

    # Kiểm tra đầu ra
    try:
        guard.validate(output)
        return f"Đầu ra: {output}"
    except Exception as e:
        return f"Đầu ra bị chặn: {e}"

# Thử nghiệm
print(guarded_openai_call("Tell me about Google products."))  # Bị chặn vì nhắc "Google"
print(guarded_openai_call("What's the weather like today?"))  # Hợp lệ
```
